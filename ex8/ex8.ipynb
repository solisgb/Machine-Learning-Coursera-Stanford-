## Part 2: Estimate the dataset statistics

```octave
whos;
```

```octave
# mean
m = size(X, 1);
n = size(X, 2);
mu = zeros(n, 1);
for i = 1:n
    mu(i, 1) = sum(X(:, i));
end
mu = mu / m;
fprintf('mu %f %f', mu);
```

```octave
%variance
x = zeros(m, n);
sigma2 = zeros(n, 1);
```

```octave
for i = 1:n
    x(:, i) = (X(:, i) - mu(i, 1)).^2;
    sigma2(i, 1) = sum(x(:, i));
end
sigma2 = sigma2 / m;
fprintf('sigma2 %f %f', sigma2);
```

```octave
visualizeFit(X,  mu, sigma2);
xlabel('Latency (ms)');
ylabel('Throughput (mb/s)');
```
